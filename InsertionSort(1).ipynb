# 1. Problema de ordenamiento

Dada una entrada: $[a1,a2,...,an]$

Obtener una salida: $[a'1,a'2,...,a'n]$

Tal que: $a_1'\leq a_2' \leq... \leq a_n'$

Uno de los problemas de la computación consiste en poder ordenar un conjunto de datos de manera efectiva, para esto han surgido múltiples algoritmos de ordenamiento, entre los cuales se encuentran: RadixSort, SelectionSort, BubbleSort y el InsertionSort. Y a pesar de la inmensa cantidad e algoritmos de ordenamiento que existen, se siguen creando nuevos hasta el día de hoy, cada uno con sus ventajas y desventajas en múltiples ámbitos y con distintas complejidades y usos de memoria.

# 2. InsertionSort

A continuación se muestra un ejemplo de como sería la implementación del algoritmo **InsertionSort**.


## 2.1. Código

```python
##jeJEje
##hacerlo teniendo en cuenta las cosas que hay que mostrar 
import random
from termcolor import colored
import copy

def insertionSort(array , verbose = False):
    contRep = 0
    length = len(array)

    if verbose == True: 
          print("input array:", array)

    for i in range(1 , length):
        if verbose == True: 
            print("\nPasada", i)
        actual = array[i]
        k = i - 1

        if verbose == True:
            print("Se revisa si ", actual, "está la posición correcta")

        while k >= 0 and actual < array[k]:
          if verbose == True:
            print(array)
            print("El elemento ", actual, "está en una posición incorrecta y se intercambia con ", array[k])

          array[k + 1] = array[k]
          k -= 1
          contRep += 1
          array[k + 1] = actual
          if verbose == True:
              print(array)
    print("El arreglo se ha ordenado")
    return array, contRep

array = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0]
print("Entrada: ", array)
array, counter = insertionSort(array)
print("Salida: ", array)
print("# comparaciones: ", counter)
  
```

## 2.2. Descripción del algoritmo

El algoritmo recibe como entrada una lista (o arreglo) $a$ con una cantidad $n$ de elementos que vamos a querer ordenar dentro de la misma lista.

1. En cada iteración $i$, el algoritmo iterará desde $a[1]$ hasta $a[N-1]$ comparando el elemento actual con el anterior, de manera que, si $a[k]$ < $a[k-1]$, los elementos cambiarán de posiciones.

2. Luego se volverá a revisar si la nueva posición del elemento es la correcta, comparando con el elemento anterior de la misma forma que antes, una vez que el elemento se encuentre en la posición correcta, se continuará con el resto del arreglo.

3. Una vez que el arreglo se encuentre ordenado el algoritmo se acaba y retorna el arreglo ordenado con el contador de intercambios.

## 2.3. Ejemplo

Si tenemos el siguiente arreglo:

$a=[4,3,6,5,1]$

Se revisa si el 3 > 4, como no es cierto, se intercambian las posiciones

$a=[3,4,6,5,1]$

Como el 3 no tiene ningún elemento antes, se pasa al siguiente, y como 6 > 4 no se hace ningún cambio:

$a=[3,4,6,5,1]$

Comparamos el siguiente elemento con su anterior, como 5 > 6 es falso, intercambiamos los elementos de lugar:

$a=[3,4,5,6,1]$

Ahora comprobamos si el 5 está en el lugar correcto revisando si 5 > 4, como esto es cierto no se hace ningún otro cambio y se pasa al siguiente elemento.

$a=[3,4,5,6,1]$

Pasamos al 1, como 1 > 6 es falso, intercambiamos los dos elementos de lugar:

$a=[3,4,5,1,6]$

Revisamos si el 1 está en la posición correcta revisando si 1 > 5, como esto vuelve a ser falso, hacemos otro cambio:

$a=[3,4,1,5,6]$

Revisamos si el 1 está en la posición correcta comprobando si 1 > 4, como esto es falso, cambiamos los elementos de lugar:

$a=[3,1,4,5,6]$

Hacemos la última revisión con 1 > 3, como es falso, hacemos otro cambio más:

$a=[1,3,4,5,6]$

Y ya tenemos el arreglo ordenado.

# 3. Tiempo de ejecución

### **Teorema (Tiempo de ejecución).**

*El algoritmo **InsertionSort** tiene un **tiempo de ejecución de** $O(n^2)$ en el peor caso.*

## Prueba del teorema

## Prueba del teorema

Si revisamos el código, podemos apreciar que el **peor caso** ocurre cuando el arreglo está ordenado de **mayor a menor**. Ya que habría que mover cada elemento al principio del arreglo. 


## Ejemplo

```python
array = [4, 3, 2, 1, 0]
print("Entrada: ", array)
array, counter = insertionSort(array,True)
print("Salida: ", array)
print("# comparaciones: ", counter)
```

En los peores casos cada elemento fue comparado $i-1$ veces, es decir fueron comparados e intercambiados la mayor cantidad de veces posible para su posición, con un total de 10 movimientos para un arreglo de 5 elementos, siendo este el peor caso posible.

En el caso general se deberán realizar $\sum\limits_{i=1}^{n} i-1$ movimientos.

# 4. Correctitud

### **Teorema (Correctitud).**

*El algoritmo **InsertionSort** genera un arreglo: $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir,* $a_1'\leq a_2' \leq... \leq a_n'$.


## Prueba del Teorema

Previamente observamos que en cada iteración de InsertionSort, si estamos en una posición i, revisamos si el elemento está en el lugar correcto.

>Al comienzo de cada iteración del bucle mayor, el sub arreglo formado por los elementos  $[a[0],...,a[i−1]]$  contiene los mismos elementos que el arreglo original  $[a_1,a_2,a_3,...,a[i−1]]$  pero ordenados de menor a mayor. Y el resto del arreglo: a[i,..,n-1] corresponde al resto de elementos que no se han ordenado aún.


**Inicialización**

Empezamos con $i = 1$, por lo tanto, a[0..i-1] corresponde al primer elemento del arreglo a[0], que está encuentra trivialmente ordenado.


**Mantención**

Para esto revismos la i-ésima iteración. Con esto asumimos que antes de comenzar la iteración, el arreglo a[0..i-1] se encuentra ordenado y contiene los elementos del arreglo original.

Durante la iteración, en el caso de que el elemento de la posición i no esté en el lugar correcto, lo “movemos” desde la posición i, hasta su posición correcta. Por lo que, después de hacer eso, el arreglo a[0..i] también contendrá los elementos del arreglo original (ya que agregamos el i-ésimo elemento) y estará ordenado.


**Correctitud**

Como se cumplen tanto la inicialización y la mantención, podemos decir que al **finalizar la n-ésima iteración del algoritmo**, se que genera un arreglo $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor.

## 5.1. Número de comparaciones

A continuación se muestra gráficamente una comparación entre:

* La cantidad de comparaciones del peor caso calculadas matemáticamente,
* La cantidad de comparaciones del mejor caso calculadas matemáticamente (n-1 comparaciones si el arreglo está ordenado)
* La cantidad de comparaciones realizadas experimentalmente para entradas aleatorias

para tamaños de problemas con $n \in [5,19]$.

Los arreglos de entrada son generados aleatoriamente haciendo un sampling de un conjunto de números entre 1 y 100.

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y1=[n*(n-1)/2 for n in range(5,20)] # worst case
y2=[n-1 for n in range(5,20)] # best case
y=[]; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a,counter = insertionSort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor Caso Teórico", "Mejor Caso Teórico"])

plt.xlabel('n')
plt.ylabel('number of operations')
plt.show()
```

### Análisis de resultados

En el gráfico se puede apreciar que, como cabía de esperar, el tiempo de ejecución empírico esta por debajo del peor caso teórico.

Y con respecto al mejor caso teórico, este apenas cambia.

## 5.2. Tiempo de ejecución (experimental)

En el siguiente gráfico se muestra el tiempo de ejecución del algoritmo **InsertionSort** en función del tamaño del problema para arreglos generados aleatoriamente de tamaños entre $5$ y $500$.



```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  t = repeat(setup="from __main__ import insertionSort", stmt=f"insertionSort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["InsertionSort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```

### Análisis de resultados

En el gráfico se puede apreciar, como era de esperarse, que el tiempo de jecución experimental se comporta de manera cuadrática en función del tamaño del problema (al igual que en el análisis teórico).
